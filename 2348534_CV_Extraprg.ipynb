```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
image = cv2.imread('/content/desktop-wallpaper-messi-of-lionel-messi-messi-cool-thumbnail.jpg')
```

```python
sampling_factor = 2
sampled_image = image[::sampling_factor, ::sampling_factor]

quantization_levels = 8
quantized_image = np.floor_divide(image, 256 // quantization_levels) * (256 // quantization_levels)
```

```python
cv2_imshow(image)
cv2_imshow(sampled_image)
cv2_imshow(quantized_image.astype(np.uint8))
```

```python
outputnew_size = (image.shape[1] // 2, image.shape[0] // 2)
interpolation_methods = [cv2.INTER_LINEAR, cv2.INTER_CUBIC, cv2.INTER_NEAREST, cv2.INTER_LANCZOS4]
method_names = ['Bi-linear', 'Bi-cubic', 'Nearest neighbor', 'LANCZOS4']

for method, name in zip(interpolation_methods, method_names):
    resized_imagenew = cv2.resize(image, outputnew_size, interpolation=method)
```

```python
cv2_imshow(resized_imagenew)
```
